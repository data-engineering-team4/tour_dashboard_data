```python
import pandas as pd

targetUrl = "https://raw.githubusercontent.com/data-engineering-team4/tour_dashboard_data/main/csv_files/analytics/%EC%A7%80%EC%97%AD%EB%B3%84_%EC%8B%9C%EC%84%A4%EB%B3%84_%EC%8B%9C%EC%84%A4%ED%86%B5%EA%B3%84.csv"
FacilitySumData = pd.read_csv(targetUrl)    # 판다스로. csv파일을 읽어라.
#FacilitySumData.to_csv(f'지역별_시설별_시설통계.csv', sep=',', na_rep='NaN',index = False)
FacilitySumData
```

```python
data_list = []
for i in FacilitySumData.area_name.unique():
    for k in FacilitySumData.facility_type.unique():
        data = sum(FacilitySumData.facility_count[(FacilitySumData.area_name==i)&(FacilitySumData.facility_type==k)])
        data_list.append([i,k,data])
print(data_list)
```

```python
area_type_df = pd.DataFrame(data_list)
area_type_df.to_csv("시도별 시설별 개수총합.csv",header=None,index=None)
```

```python
import pandas as pd

targetUrl = "https://raw.githubusercontent.com/data-engineering-team4/tour_dashboard_data/main/csv_files/AREA_TOUR_FACILITY.csv"
FacilityData = pd.read_csv(targetUrl)    # 판다스로. csv파일을 읽어라.
FacilityData
```

```python
import pandas as pd

targetUrl = "https://raw.githubusercontent.com/data-engineering-team4/tour_dashboard_data/main/csv_files/CITY_CODE.csv"
City_Code_Data = pd.read_csv(targetUrl)    # 판다스로. csv파일을 읽어라.
City_Code_Data.rename(columns={'PARENT_CODE':'AREACODE'},inplace=True)
City_Code_Data.rename(columns={'CODE':'SIGUNGUCODE'},inplace=True)
City_Code_Data
```

```python
import pandas as pd

targetUrl = "https://raw.githubusercontent.com/data-engineering-team4/tour_dashboard_data/main/csv_files/CODE_DETAIL.csv"
Code_Detail_Data = pd.read_csv(targetUrl)    # 판다스로. csv파일을 읽어라.
Code_Detail_Data
```

```python
Area_Data = Code_Detail_Data[Code_Detail_Data.CODE_CATEGORY=='AREA']
Area_Data = Area_Data.drop(['PARENT_CODE '],axis=1)
Area_Data.rename(columns={'CODE':'AREACODE'},inplace=True)
Area_Data
```

```python
Area_merge_df = pd.merge(FacilityData, Area_Data, on='AREACODE')
#Area_merge_df = merge_df.drop(['CODE_CATEGORY'],axis=1)
Area_merge_df
```

```python
merge_df = pd.merge(Area_merge_df, City_Code_Data, on=['AREACODE','SIGUNGUCODE'])
merge_df = merge_df.drop(['Unnamed: 0','CODE_CATEGORY_x','CODE_CATEGORY_y'],axis=1)
merge_df.rename(columns={'CODE_NM_x':'AREA'},inplace=True)
merge_df.rename(columns={'CODE_NM_y':'CITY'},inplace=True)
merge_df= merge_df.drop(['CATEGORY','CODE_SEQ'],axis=1)
```
