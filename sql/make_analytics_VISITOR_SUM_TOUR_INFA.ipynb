```python
%%sql

SELECT B.CODE_SEQ
     , A.CODE_NM  AREA_NM
     , A.CODE     AREA_CD
  FROM RAW_DATA.CODE_DETAIL A
     , RAW_DATA.CODE_DETAIL B 
 WHERE A.CODE = B.PARENT_CODE
   AND A.CODE_CATEGORY = 'AREA'
   AND B.CODE_CATEGORY = 'CITY';
```

```python
%%sql

CREATE TABLE ANALYTICS.VISITOR_SUM AS
WITH AREA AS (
  SELECT B.CODE_SEQ
     , A.CODE_NM  AREA_NM
     , A.CODE     AREA_CD
  FROM RAW_DATA.CODE_DETAIL A
     , RAW_DATA.CODE_DETAIL B 
 WHERE A.CODE = B.PARENT_CODE
   AND A.CODE_CATEGORY = 'AREA'
   AND B.CODE_CATEGORY = 'CITY'
)
SELECT B.AREA_NM 
     , TO_CHAR(A.VISIT_DT, 'YYYY-MM-DD')  VISIT_DT
     , SUM(A.TOURIST_NM)                  VISITOR_SUM
     , CASE WHEN A.TOURIST_CD IN ('1', '2')
            THEN '내국인'
            ELSE '외국인' END             TOURIST_TYPE 
  FROM RAW_DATA.VISITOR_SESSION A
  JOIN AREA B
    ON A.CODE_SEQ = B.CODE_SEQ 
 GROUP BY 1, 2, 4
 ORDER BY 1, 2, 4
```

```python
%%sql

INSERT INTO ANALYTICS.VISITOR_SUM
(
WITH AREA AS (
  SELECT B.CODE_SEQ
     , A.CODE_NM  AREA_NM
     , A.CODE     AREA_CD
  FROM RAW_DATA.CODE_DETAIL A
     , RAW_DATA.CODE_DETAIL B 
 WHERE A.CODE = B.PARENT_CODE
   AND A.CODE_CATEGORY = 'AREA'
   AND B.CODE_CATEGORY = 'CITY'
)
SELECT B.AREA_NM 
     , TO_CHAR(A.VISIT_DT, 'YYYY-MM-DD')  VISIT_DT
     , SUM(A.TOURIST_NM)                  VISITOR_SUM
     , '모든'                      TOURIST_TYPE 
  FROM RAW_DATA.VISITOR_SESSION A
  JOIN AREA B
    ON A.CODE_SEQ = B.CODE_SEQ 
 GROUP BY 1, 2
 ORDER BY 1, 2
)
```

```python
%%sql


CREATE TABLE  ANALYTICS.TOUR_INFRA_CNT AS 
SELECT B.CODE_NM             AREA_NM
     , COUNT(CASE WHEN A.FACILITY_TYPE_CD = '32' 
                  THEN 1 END)                       ACCOMMODATION_CNT 
     , COUNT(CASE WHEN A.FACILITY_TYPE_CD = '38' 
                  THEN 1 END)                       SHOPPING_CNT 
     ,  COUNT(CASE WHEN A.FACILITY_TYPE_CD = '39' 
                  THEN 1 END)                       RESTAURANT_CNT             
  FROM RAW_DATA.AREA_TOUR_FACILITY A
  JOIN RAW_DATA.CODE_DETAIL B
    ON A.AREACODE = B.CODE
 WHERE B.CODE_CATEGORY = 'AREA'
GROUP BY 1;
```

```python
result = %sql SELECT * FROM ANALYTICS.TOUR_INFRA_MATRIX;
df = result.DataFrame()
df.to_csv('TOUR_INFRA_MATRIX.csv', index=False)
```

```python
%%sql

CREATE TABLE ANALYTICS.TOUR_INFRA_MATRIX AS 
SELECT B.CODE_NM             AREA_NM
     , CASE WHEN A.FACILITY_TYPE_CD = '32' 
            THEN '숙소'
            WHEN A.FACILITY_TYPE_CD = '38'
            THEN '쇼핑'
            WHEN A.FACILITY_TYPE_CD = '39'
            THEN '음식점' END                      METRIC
     , CASE WHEN A.FACILITY_TYPE_CD = '32'
            THEN C.ACCOMMODATION_CNT
            WHEN A.FACILITY_TYPE_CD = '38'
            THEN C.SHOPPING_CNT
            WHEN A.FACILITY_TYPE_CD = '39'
            THEN C.RESTAURANT_CNT END             "VALUE"          
  FROM RAW_DATA.AREA_TOUR_FACILITY A
  JOIN RAW_DATA.CODE_DETAIL B
    ON A.AREACODE = B.CODE
  JOIN ANALYTICS.TOUR_INFRA_CNT C 
    ON B.CODE_NM = C.AREA_NM
 WHERE B.CODE_CATEGORY = 'AREA'
   AND METRIC IS NOT NULL
GROUP BY 1, 2, 3
ORDER BY 1, 2, 3;
```
